```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([0,-46,-12,-15.5,-30.5,-30])
y1 = np.array([0,-17,11,12,2,18.5])
x2 = np.array([115,72,106,101,85,86])
y2 = np.array([0,-23,11,10,-3,3.5])
d = 35
theta = np.arctan2(y2 - y1, x2 - x1)
norms = np.sqrt((x2-x1)**2+(y2-y1)**2)
xq = (x1+x2)/2 - d*(y2-y1)/norms
yq = (y1+y2)/2 + d*(x2-x1)/norms
xe = xq - xq[0]
ye = yq - yq[0]
e = np.sqrt(xe**2+ye**2)
print(xq)
print(yq)
print("dx mean: %f std: %f" % (xe[1:].mean(), xe[1:].std()))
print("dy mean: %f std: %f" % (ye[1:].mean(), ye[1:].std()))
print("dtheta mean: %f std: %f" % (theta[1:].mean()*180/np.pi, theta[1:].std()*180/np.pi))

print("dx mean: %f std: %f" % (xe[1:].mean()/35.56, xe[1:].std()/35.56))
print("dy mean: %f std: %f" % (ye[1:].mean()/35.56, ye[1:].std()/35.56))
print("dtheta mean: %f std: %f" % (theta[1:].mean()*180/np.pi/35.56, theta[1:].std()*180/np.pi/35.56))
```

```python


for i in range(1,5):
    plt.plot((x1[i], x2[i]), (y1[i], y2[i]),'.--', color='k')
    plt.plot(((x1[i]+x2[i])/2, xq[i]), ((y1[i]+y2[i])/2, yq[i]), '.-', color="k")
i = 5
plt.plot((x1[i], x2[i]), (y1[i], y2[i]),'.--', color='k', label="Final position")
plt.plot(((x1[i]+x2[i])/2, xq[i]), ((y1[i]+y2[i])/2, yq[i]), '.-', color="k")
i = 0
plt.plot
plt.plot((x1[i], x2[i]), (y1[i], y2[i]), '.-', color="b", label="Starting position")
plt.plot(((x1[i]+x2[i])/2, xq[i]), ((y1[i]+y2[i])/2, yq[i]), '.-', color="b")
plt.axis("equal")
plt.grid()
plt.show()
```
