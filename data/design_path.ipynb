```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
from matplotlib.patches import Rectangle
```

```python
x1lap = [1.0, 0.75, 0.4, 0.4, 0.75, 0.75, 0.4, 0.4, 0.75, 1.0,  0.8, 0.4, 0.75, 1.25, 1.6, 1.2,  1.0, 1.25, 1.6, 1.5]
y1lap = [0.4, 0.8, 0.85, 1.2, 1.25, 1.6, 1.65, 2.0, 2.05, 2.15, 2.4, 2.4, 2.05, 2.05, 2.4, 2.4, 2.15, 2.05, 1.5, 0.5] 
n1lap = len(x1lap)
print(n1lap)
nlaps = 5
nc = n1lap*nlaps+1
qc = np.arange(0, nc)
q = np.linspace(0,nc-1, 4000)
xc = np.zeros(nc)
yc = np.zeros(nc)
for i in range(nc):
    xc[i] = x1lap[i % n1lap]
    yc[i] = y1lap[i % n1lap]

xref = CubicSpline(qc, xc)(q)
yref = CubicSpline(qc, yc)(q)

fig, ax = plt.subplots()
ax.add_patch(Rectangle((0,0), 3.0, 2.0, edgecolor="k",fill=False))
ax.plot(yref, xref)
# ax.scatter(yc, xc)
ax.axis("equal")
ax.set_ylim(ax.get_ylim()[::-1])
ax.grid()
ax.set_xlabel("y [m]")
ax.set_ylabel("x [m]")
fig.tight_layout()
plt.savefig("path-design.png")
plt.savefig("path-design.pdf")
plt.show()
```
